## Dataset:
https://www.kaggle.com/datasets/uom190346a/sleep-health-and-lifestyle-dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sleep_df = pd.read_csv("SleepHealthAndLifestyleDataset.csv")
```

```python
sleep_df
```

### Getting the data types for each column:

```python
sleep_df.dtypes
```

### We need to change Sleep Disorder column (categorical) to have numerical values.
* We need to change NaN (null) values to 0 - the absence of a sleeping disorder
* Sleep Apnea and Insomnia will be changed to 1 - the presence of a sleeping disorder

```python
sleep_df.isna().sum()
```

**We can see that the only column with null values is Sleep Disorder. There are 219 null values, but in our case, they represent the absence of a sleeping disorder**

```python
sleep_df['Sleep Disorder'] = sleep_df['Sleep Disorder'].fillna(0)
```

```python
sleep_df
```

```python
sleep_df['Sleep Disorder'].replace('Sleep Apnea', 1, inplace=True)
```

```python
sleep_df['Sleep Disorder'].replace('Insomnia', 1, inplace=True)
```

```python
sleep_df
```

```python
sleep_df.dtypes
```

**The Sleep Disorder column now only contains numerical values. 0 - absence of sleep disorder; 1 - presence of a sleep disorder**

### Visualizations

#### Occupation vs. Sleep Disorder

```python
# count plot
plt.figure(figsize=(16,6))
sns.countplot(x='Occupation', hue='Sleep Disorder', data = sleep_df)
plt.title("Distribution of Presence of Sleep Disorder by Occupation")
plt.xlabel('Occupation')
plt.ylabel('Sleep Disorder')
plt.legend(title='Sleep Disorder Presence', labels=['Absent', 'Present'])
plt.show()
```

##### Pre-Processing the BMI Category column
* "Normal Weight" should be changed into "normal"

```python
sleep_df['BMI Category'].replace('Normal Weight', 'Normal', inplace = True)
```

#### BMI Category vs. Sleep Disorder

```python
# heat map
pivot_table = sleep_df.pivot_table(index='BMI Category', columns='Sleep Disorder', aggfunc='size', fill_value=0)
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_table, annot=True, fmt = 'd', cmap='coolwarm')
plt.title('Correlation between BMI Category and Sleep Disorder Presence')
plt.xlabel('Sleep Disorder Presence')
plt.ylabel('BMI Category')
plt.show()
```

#### Stress Level vs. Sleep Disorder

```python
# Bar plot: showing average (mean) stress level for the presence and absence of a sleep disorder
plt.figure(figsize=(10, 6))
sns.barplot(x='Sleep Disorder', y='Stress Level', data=sleep_df, estimator=np.mean)
plt.title('Distribution of Stress Level by Sleep Disorder Presence')
plt.xlabel('Sleep Disorder Presence')
plt.ylabel('Stress Level')
plt.show()
```

### Visualizations to learn something else from the data
* Relationship between Physical Activity Levels and BMI Category

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x="BMI Category", y="Physical Activity Level", data=sleep_df)
plt.title("Physical Activity Levels by BMI Category")
plt.xlabel("BMI Category")
plt.ylabel("Physical Activity Level by Minutes Per Day")
plt.show()
```
