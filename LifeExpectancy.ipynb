### Data Pre-Processing

```python
#Removing columns that are not necessary for this project
lifeExpectancy.isna().sum()
```

##### For the first experiment, the targer variable will be Life expectancy and the predictor variables will be BMI and Adult Mortality.
* We will need to manage the null values in the BMI and Adult Mortality columns

```python
rows_with_null = lifeExpectancy[lifeExpectancy[' BMI '].isnull() | lifeExpectancy['Adult Mortality'].isnull()]
rows_with_null
```

```python
#Drop the rows where Adult Mortality AND Life Expectancy are both null
lifeExpectancy = lifeExpectancy.dropna(subset=['Life expectancy ', 'Adult Mortality'], how='all')
```

```python
lifeExpectancy.isna().sum()
```

```python
filtered = lifeExpectancy[(lifeExpectancy['Country'].isin(['Sudan', 'South Sudan'])) & (lifeExpectancy[' BMI '].isnull())]
filtered
```

```python
filtered.shape
```

```python
#Drop rows where BMI is null. All these rows are from the coutry of Sudan and South Sudan. Since it is only 2 countries, we can remove it
#for the sake of the experiement.
lifeExpectancy = lifeExpectancy.dropna(subset=[' BMI '])
```

```python
lifeExpectancy.isna().sum()
```

```python
rows_with_null = lifeExpectancy[lifeExpectancy['Income composition of resources'].isnull() & lifeExpectancy['Schooling'].isnull()]
rows_with_null
```

```python
rows_with_null.shape
```

```python
#Drop rows where Income composition of resources AND Schooling is null. We are not predicting life expectancy based on a specific country
#therefore, we will delete these rows.
lifeExpectancy.dropna(subset=['Income composition of resources', 'Schooling'], inplace=True)
```

```python
lifeExpectancy.isna().sum()
```

### Experiment 1 - Linear Regression (Modeling and Evaluation)

```python
#Dividing data into X and y
X = lifeExpectancy[[' BMI ', 'Adult Mortality']]
y = lifeExpectancy['Life expectancy ']
```

```python
#Dividing data into testing and training data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.20, random_state=None)
```

```python
#feature scaling using standard scalar
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#Importing linear regression model from sklearn
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
#training the model
regressor = lin_reg.fit(X_train, y_train)
#making predictions on the test set
y_pred = regressor.predict(X_test)
#evaluating model performance
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

* Predicting a single value:

```python
newLifeExpectancydf = lifeExpectancy[[' BMI ', 'Adult Mortality']]
```

```python
#Data to predict life expectancy
data = np.array([[50,300]])
```

```python
scaled = sc.transform(data)
predictedLifeExp = regressor.predict(scaled)

print("Predicted Life expectancy:", predictedLifeExp)
```

### Experiment 2 - Random Forest (Modeling and Evaluation)

* We will use the same features as Experiment 1

```python
#Importing Random Forest model from sklearn
from sklearn.ensemble import RandomForestRegressor
rf_reg = RandomForestRegressor(random_state=42, n_estimators=200)
#training the model
regressor = rf_reg.fit(X_train, y_train)
#making predicitons on the test set
y_pred = regressor.predict(X_test)
#evaluating the model performance
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

* Note: The MSE decreased from the first Experiment

### Experiment 3 - Random Forest (Modeling and Evaluation)
* We will use the same algorithm as Experiment 2, Random Forest, but we will use the 'Income compositions of resources' and 'Schooling' columns for our predictor variables this time.

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,  test_size=0.20, random_state=None)
```

```python
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
rf_reg = RandomForestRegressor(random_state=42, n_estimators=200)
#training the model
regressor = rf_reg.fit(X_train, y_train)
#making predicitons on the test set
y_pred = regressor.predict(X_test)
#evaluating the model performance
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
