## Dataset
https://www.kaggle.com/datasets/ruchi798/housing-prices-in-metropolitan-areas-of-india?select=Bangalore.csv

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Bangalore_df = pd.read_csv('Bangalore.csv')
```

```python
Bangalore_df
```

```python
# Looking at the datatypes for each column
Bangalore_df.dtypes
```

```python
# Checking for null values
Bangalore_df.isna().sum()
```

```python
# Checking for duplicate rows
Bangalore_df.duplicated().sum()
```

```python
# Dropping all duplicate rows
Bangalore_df = Bangalore_df.drop_duplicates()
Bangalore_df
```

#### Crucial preprocessing steps
* Converting prices from Indian Rupees to USD - current exchange rate: 1 USD = 83.10 INR

```python
Bangalore_df['Price'] = (Bangalore_df['Price']/83.10).round()
Bangalore_df
```

### Data Understanding

```python
Bangalore_df.info()
```

```python
Bangalore_df.columns
```

```python
numeric_cols = Bangalore_df.select_dtypes(include=['number'])
numeric_cols.corr()
```

#### Based on the above correlation matrix, most of the features of the house don't have a high correlation with the price. Therefore, we can delete some of the columns in order to clean up the data to better fit the model

```python
Bangalore_df = Bangalore_df.drop(columns=['RainWaterHarvesting', 'IndoorGames', 'ShoppingMall', 'School', 'StaffQuarter', 'MultipurposeRoom', 'Hospital', 'VaastuCompliant'])
```

```python
Bangalore_df
```

### Since we want to predict house price, we need to split our price column

```python
Bangalore_df['Price_bins']=pd.cut(Bangalore_df.Price, 3)
```

```python
Bangalore_df
```

### Data Visualization

```python
freqTable = Bangalore_df['Price_bins'].value_counts()
freqTable
```

```python
freqTable.plot(kind='bar', figsize=(10,30))
plt.title('Frequency of Price Categories')
plt.xlabel('Category')
plt.ylabel('Frequency')
plt.show()
```

### Modeling our Data

```python
X = Bangalore_df[['Area', 'No. of Bedrooms', 'Resale']]
y = Bangalore_df['Price_bins']
```

```python
y.dtype
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Import classifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#converts the categories to integer codes before fitting the classifier
y_train_codes = y_train.cat.codes
```

```python
#converts the categories to integer codes before fitting the classifier
y_test_codes = y_test.cat.codes
```

```python
knn = KNeighborsClassifier(n_neighbors = 3)
```

```python
#Fit the model on training data
knn.fit(X_train,y_train_codes)
```

```python
knn.score(X_test, y_test_codes)
```

```python
predicted = knn.predict(X_test)
```

```python
np.set_printoptions(threshold=np.inf)
predicted
```

```python
from sklearn import tree
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
clf = clf.fit(X_train, y_train_codes)
```

```python
clf.score(X_test, y_test_codes)
```

```python
predicted = clf.predict(X_test)
```

```python
np.set_printoptions(threshold=np.inf)
predicted
```

### Creating a tree

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1, figsize = (6,6), dpi=300)
tree.plot_tree(clf, max_depth = 3, feature_names = X.columns.to_list(), filled=True)
plt.show()
```

```python
fi = clf.feature_importances_ #feature importance array
fi = pd.Series(data = fi, index = X.columns) #convert to Pandas series for plotting
fi.sort_values(ascending=False, inplace=True) #sort descending
```

```python
#create bar plot
plt.figure(figsize=(12, 6))
chart = sns.barplot(x=fi, y=fi.index, palette=sns.color_palette("BuGn_r", n_colors=len(fi)))
chart.set_xticklabels(chart.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```
